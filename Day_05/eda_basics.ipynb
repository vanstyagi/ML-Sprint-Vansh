```python
# EDA - Day 05: Basic Overview

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Sample DataFrame
df = pd.DataFrame({
    'Age': [22, 25, 27, 29, 24, 30, 23],
    'Salary': [25000, 40000, 42000, 50000, 32000, 55000, 30000]
})

# Describe data
df.describe()
```

```python
# Distribution Plot for Age
sns.histplot(df['Age'], kde=True)
plt.title("Age Distribution")
plt.show()

# Distribution Plot for Salary
sns.histplot(df['Salary'], kde=True, color='orange')
plt.title("Salary Distribution")
plt.show()

# Correlation Heatmap
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Grouping Example (Assume Category Column)
df['Category'] = ['A', 'B', 'A', 'B', 'B', 'A', 'A']
print(df.groupby('Category')['Salary'].mean())

# Value Counts
print("\nCategory Distribution:")
print(df['Category'].value_counts())

# Unique Values in Age
print("\nUnique Ages:")
print(df['Age'].unique())
```
